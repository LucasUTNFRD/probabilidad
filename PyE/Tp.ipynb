```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
df = pd.read_csv('/home/lucas/Downloads/afa_2015_2022_spa.csv')
df
```

```python
filtro = df['torneo'] == 'Campeonato 2022'
```

```python
df[filtro]
```

```python
df_filtrada=df[filtro]
```

```python
df_filtrada[['torneo','fecha','equipo_local','equipo_visitante','goles_local', 'goles_visitante']]
```

```python
goles_locales = df_filtrada[['equipo_local', 'goles_local']].rename(columns={'Equipo Local': 'Equipo', 'Goles Local': 'Goles'})
goles_locales.columns = ['Equipo', 'Goles']
goles_locales
```

```python
goles_visitantes = df_filtrada[['equipo_visitante', 'goles_visitante']].rename(columns={'Equipo Visitante': 'Equipo', 'Goles Visitante': 'Goles'})
goles_visitantes.columns = ['Equipo', 'Goles']
goles_visitantes
```

```python
goles_totales = pd.concat([goles_locales, goles_visitantes], ignore_index=True)
goles_totales
```

```python
goles_por_equipo = goles_totales.groupby('Equipo')['Goles'].sum().reset_index()
goles_por_equipo
```

```python
goles_por_equipo['Partidos Jugados'] = 27
goles_por_equipo
```

```python
goles_por_equipo['Promedio Goles por Partido'] = goles_por_equipo['Goles'] / goles_por_equipo['Partidos Jugados']
goles_por_equipo
```

```python
goles_por_equipo.sort_values(by='Promedio Goles por Partido',ascending=True).round(2)
```

```python
plt.figure(figsize=(8, 6))  # Define el tamaño de la figura (opcional)

# Crea el gráfico de caja
plt.boxplot(goles_por_equipo['Promedio Goles por Partido'])

# Agrega etiquetas y título
plt.ylabel('Promedio de Goles por Partido')
plt.title('Gráfico de Caja de Promedio de Goles por Partido')

# Muestra el gráfico
plt.show()
```

```python
# Define el número de intervalos (puedes ajustarlo según tus preferencias)
num_intervalos = int(np.sqrt(len(goles_por_equipo)))

# Crea el histograma
plt.figure(figsize=(10, 6))
plt.hist(goles_por_equipo['Promedio Goles por Partido'], bins=num_intervalos, edgecolor='k')

# Añade etiquetas y título
plt.xlabel('Promedio de Goles por Partido')
plt.ylabel('Frecuencia Absoluta')
plt.title('Histograma de Promedio de Goles por Partido')

# Calcula los límites de los intervalos
hist, bin_edges = np.histogram(goles_por_equipo['Promedio Goles por Partido'], bins=num_intervalos)

# Calcula la frecuencia relativa
frecuencia_relativa = hist / len(goles_por_equipo)

# Calcula la frecuencia acumulada
frecuencia_acumulada = np.cumsum(hist)

# Calcula la frecuencia relativa acumulada
frecuencia_relativa_acumulada = frecuencia_acumulada / len(goles_por_equipo)

# Crea una tabla con los datos
tabla_histograma = pd.DataFrame({
    'Límite Inferior': bin_edges[:-1],
    'Límite Superior': bin_edges[1:],
    'Frecuencia Absoluta': hist,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Acumulada': frecuencia_acumulada,
    'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
})

# Muestra la tabla
tabla_histograma.round(2)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.axis('tight')
ax.axis('off')
goles_por_equipo=goles_por_equipo.sort_values(by='Promedio Goles por Partido',ascending=True)
tabla = ax.table(cellText=goles_por_equipo.values, colLabels=goles_por_equipo.columns, cellLoc='center', loc='center')
tabla.auto_set_font_size(True)
tabla.set_fontsize(16)
```

```python
# Datos para los cuales deseas calcular los cuartiles
data = goles_por_equipo['Promedio Goles por Partido']

# Calcula los cuartiles
q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)  # Este es el cuartil 2 (mediana)
q3 = np.percentile(data, 75)

print("Q1:", q1)
print("Q2 (Mediana):", q2)
print("Q3:", q3)
```

```python
# Define el número de intervalos (puedes ajustarlo según tus preferencias)
num_intervalos = int(np.sqrt(len(goles_por_equipo)))

# Crea el histograma
plt.figure(figsize=(10, 6))
plt.hist(goles_por_equipo['Promedio Goles por Partido'], bins=num_intervalos, edgecolor='k')

# Añade un polígono de frecuencia
frecuencia, bordes = np.histogram(goles_por_equipo['Promedio Goles por Partido'], bins=num_intervalos)
centros = (bordes[:-1] + bordes[1:]) / 2
plt.plot(centros, frecuencia, '-o', label='Polígono de Frecuencia')

# Calcula la frecuencia acumulada relativa
frecuencia_acumulada_relativa = np.cumsum(frecuencia) / len(goles_por_equipo)

# Crea una ojiva de frecuencia acumulada relativa
plt.figure(figsize=(10, 6))
plt.plot(centros, frecuencia_acumulada_relativa, '-o', label='Ojiva de Frecuencia Acumulada Relativa')
plt.xlabel('Valores')
plt.ylabel('Frecuencia Acumulada Relativa')
plt.legend()

# Crea un gráfico de torta a partir de los intervalos del histograma
plt.figure(figsize=(8, 8))
plt.pie(frecuencia, labels=[f'Intervalo {i+1}' for i in range(num_intervalos)], autopct='%1.1f%%', startangle=140)
plt.title('Gráfico de Torta de los Intervalos del Histograma')

# Muestra los gráficos
plt.show()
```

```python
# Promedio
promedio = goles_por_equipo['Promedio Goles por Partido'].mean()
print("Promedio (Media):", round(promedio,2))

# Desvío Estándar
desvio_estandar = np.std(goles_por_equipo['Promedio Goles por Partido'])
print("Desvío Estándar:", desvio_estandar)

desvio_normal = np.std(goles_por_equipo['Promedio Goles por Partido'], ddof=1)
print("Desvío Normal:", desvio_normal)

# Asimetría
mediana = goles_por_equipo['Promedio Goles por Partido'].median()
asimetria = 3 * (promedio - mediana) / desvio_estandar
print("Asimetría:", round(asimetria,2))

# Curtosis
curtosis = ((goles_por_equipo['Promedio Goles por Partido'] - promedio) ** 4).mean() / (desvio_estandar ** 4) - 3
print("Curtosis:", round(curtosis,2))

# Moda
moda = goles_por_equipo['Promedio Goles por Partido'].mode().values[0]
print("Moda:", round(moda,2))

# Cuartiles
q1 = goles_por_equipo['Promedio Goles por Partido'].quantile(0.25)
q3 = goles_por_equipo['Promedio Goles por Partido'].quantile(0.75)
print("Cuartil 1 (Q1):", round(q1,2))
print("Cuartil 3 (Q3):", round(q3,2))
```

```python
frecuencia_valores = goles_por_equipo['Promedio Goles por Partido'].round(2).value_counts()

# Encuentra el valor máximo de frecuencia
frecuencia_maxima = frecuencia_valores.max()

# Filtra los valores que tienen la frecuencia máxima
modas = frecuencia_valores[frecuencia_valores == frecuencia_maxima]

# Muestra las modas
print("Modas:", modas.index.tolist())
```

```python
promedio_goles = goles_por_equipo['Promedio Goles por Partido']
error_estandar = np.std(promedio_goles) / np.sqrt(len(promedio_goles))

error_estandar
```

```python
confianza_90 = stats.t.interval(0.9, len(promedio_goles) - 1, loc=np.mean(promedio_goles), scale=stats.sem(promedio_goles))
confianza_95 = stats.t.interval(0.95, len(promedio_goles) - 1, loc=np.mean(promedio_goles), scale=stats.sem(promedio_goles))
confianza_99 = stats.t.interval(0.99, len(promedio_goles) - 1, loc=np.mean(promedio_goles), scale=stats.sem(promedio_goles))
```
